```python
import matplotlib.pyplot as plt

names = ['Arystan', 'Tima', 'Yersultan', 'Kuanysh', 'Ruslan']
scores = [98, 37, 88, 77, 100]
plt.figure(figsize=(8, 4))
plt.plot(names, scores, marker='o', color='b', linestyle='-')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.title('Student Scores Trend')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.bar(names, scores, color='lightgreen')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.title('Student Names vs Scores')
plt.show()
```
