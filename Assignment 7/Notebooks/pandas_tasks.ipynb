```python
import pandas as pd

d = {
    'Name': ['Arystan', 'Tima', 'Yersultan', 'Kuanysh', 'Ruslan'],
    'Age': [100, 10, 22, 25, 2],
    'Score': [98, 67, 88, 77, 100]
}
df = pd.DataFrame(d)
print(df)
```

```python
print("\nFirst 3 rows:\n", df.head(3))
print("\nSummary Statistics:\n", df.describe())
print("\nAverage Score:", df['Score'].mean())
```

```python
GF = df[df['Age'] > 20]
HS = df[df['Score'] > 80]

print(GF, "\n")
print(HS)
```
